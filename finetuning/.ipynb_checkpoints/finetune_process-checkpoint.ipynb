```python
import os
import pandas as pd
```

```python
file_list = os.listdir('./lab wiki')
file_list
```

```python
csv = pd.read_csv('lab_wiki.csv')
drop_csv = csv.drop(index=[1, 3], axis='index')
drop_csv
# csv[csv['생성자'].str.contains('')]
```

```python
qna = []

titles = sorted(drop_csv.이름.tolist(), key=str.lower)
```

```python
for title, file in zip(titles, file_list):
    with open(f'./lab wiki/{file}', encoding='utf-8') as f:
        qna.append((title, '\n'.join([line.strip() for line in f.readlines()[5:] if line.strip()])))
```
